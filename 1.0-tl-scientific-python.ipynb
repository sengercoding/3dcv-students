## Task 3

Plot the color histogram of the rgb image.

Reference: https://matplotlib.org/users/index.html

````python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt

def plot_color_histogram(image):
    '''
    Plot the histogram of each color channel into a single plot.

    image -- RGB image
    '''
    # Compute the histogram for each color channel
    r_hist, r_bins = np.histogram(image[:, :, 0], bins=256, range=(0, 256))
    g_hist, g_bins = np.histogram(image[:, :, 1], bins=256, range=(0, 256))
    b_hist, b_bins = np.histogram(image[:, :, 2], bins=256, range=(0, 256))
    
    # Plot the histograms
    plt.figure(figsize=(10, 6))
    plt.plot(r_bins[:-1], r_hist, color='red', alpha=0.5, label='Red channel')
    plt.plot(g_bins[:-1], g_hist, color='green', alpha=0.5, label='Green channel')
    plt.plot(b_bins[:-1], b_hist, color='blue', alpha=0.5, label='Blue channel')

    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.title('Color Histogram of RGB Image')
    plt.legend()
    plt.show()

# Load the image
image = io.imread('./data/pepo.jpg')

# Plot the color histogram
plot_color_histogram(image)
````

## Task 4

Convert the RGB image into the HSV color space. Rotate the colors in the HSV color space, convert it back to the RGB color space and plot it together with its histogram. For this task, you are only allowed to use NumPy.

More information about the HSV color space: https://en.wikipedia.org/wiki/HSL_and_HSV

````python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

def rgb2hsv(image):
    '''
    Convert the rgb image into hsv color space and return it.

    image -- rgb image
    '''
    image = image / 255.0
    r, g, b = image[:, :, 0], image[:, :, 1], image[:, :, 2]
    maxc = np.max(image, axis=2)
    minc = np.min(image, axis=2)
    v = maxc
    s = (maxc - minc) / maxc
    s[maxc == 0] = 0
    rc = (maxc - r) / (maxc - minc)
    gc = (maxc - g) / (maxc - minc)
    bc = (maxc - b) / (maxc - minc)
    h = np.zeros_like(maxc)
    h[r == maxc] = bc[r == maxc] - gc[r == maxc]
    h[g == maxc] = 2.0 + rc[g == maxc] - bc[g == maxc]
    h[b == maxc] = 4.0 + gc[b == maxc] - rc[b == maxc]
    h = (h / 6.0) % 1.0
    hsv = np.stack([h, s, v], axis=2)
    return hsv

def hsv2rgb(image):
    '''
    Convert the hsv image into rgb color space and return it.

    image -- hsv image
    '''
    h, s, v = image[:, :, 0], image[:, :, 1], image[:, :, 2]
    i = (h * 6.0).astype(int)
    f = (h * 6.0) - i
    p = v * (1.0 - s)
    q = v * (1.0 - f * s)
    t = v * (1.0 - (1.0 - f) * s)
    i = i % 6
    conditions = [i == 0, i == 1, i == 2, i == 3, i == 4, i == 5]
    rgb = np.zeros_like(image)
    rgb[conditions[0]] = np.stack([v, t, p], axis=2)[conditions[0]]
    rgb[conditions[1]] = np.stack([q, v, p], axis=2)[conditions[1]]
    rgb[conditions[2]] = np.stack([p, v, t], axis=2)[conditions[2]]
    rgb[conditions[3]] = np.stack([p, q, v], axis=2)[conditions[3]]
    rgb[conditions[4]] = np.stack([t, p, v], axis=2)[conditions[4]]
    rgb[conditions[5]] = np.stack([v, p, q], axis=2)[conditions[5]]
    return (rgb * 255).astype(np.uint8)

def plot_color_histogram(image):
    '''
    Plot the histogram of each color channel into a single plot.

    image -- RGB image
    '''
    r_hist, r_bins = np.histogram(image[:, :, 0], bins=256, range=(0, 256))
    g_hist, g_bins = np.histogram(image[:, :, 1], bins=256, range=(0, 256))
    b_hist, b_bins = np.histogram(image[:, :, 2], bins=256, range=(0, 256))
    
    plt.figure(figsize=(10, 6))
    plt.plot(r_bins[:-1], r_hist, color='red', alpha=0.5, label='Red channel')
    plt.plot(g_bins[:-1], g_hist, color='green', alpha=0.5, label='Green channel')
    plt.plot(b_bins[:-1], b_hist, color='blue', alpha=0.5, label='Blue channel')

    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.title('Color Histogram of RGB Image')
    plt.legend()
    plt.show()

# Load the image
image = io.imread('./data/pepo.jpg')

# Convert RGB to HSV
hsv_image = rgb2hsv(image)

# Rotate the colors in the HSV color space by modifying the hue channel
hsv_image[:, :, 0] = (hsv_image[:, :, 0] + 0.5) % 1.0

# Convert HSV back to RGB
rotated_image = hsv2rgb(hsv_image)

# Plot the original and modified RGB images along with their histograms
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

axes[0, 0].imshow(image)
axes[0, 0].set_title('Original RGB Image')
axes[0, 0].axis('off')

axes[0, 1].imshow(rotated_image)
axes[0, 1].set_title('Rotated RGB Image')
axes[0, 1].axis('off')

axes[1, 0].hist(image[:, :, 0].ravel(), bins=256, color='red', alpha=0.5, label='Red channel')
axes[1, 0].hist(image[:, :, 1].ravel(), bins=256, color='green', alpha=0.5, label='Green channel')
axes[1, 0].hist(image[:, :, 2].ravel(), bins=256, color='blue', alpha=0.5, label='Blue channel')
axes[1, 0].set_title('Original Image Histogram')
axes[1, 0].legend()

axes[1, 1].hist(rotated_image[:, :, 0].ravel(), bins=256, color='red', alpha=0.5, label='Red channel')
axes[1, 1].hist(rotated_image[:, :, 1].ravel(), bins=256, color='green', alpha=0.5, label='Green channel')
axes[1, 1].hist(rotated_image[:, :, 2].ravel(), bins=256, color='blue', alpha=0.5, label='Blue channel')
axes[1, 1].set_title('Rotated Image Histogram')
axes[1, 1].legend()

plt.show()
````
